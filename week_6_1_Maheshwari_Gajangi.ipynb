```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/train.csv')
```

```python
df = pd.DataFrame(data)
```

```python
for x in df.columns:
  if df[x].dtypes == 'O':
    df[x] = df[x].str.lower().str.replace('-','_').str.replace(' ','_')
```

```python
df['Sales'] = df['Sales'].round(2)
```

###3. Data Cleaning

### 5. Histograms

```python
plt.figure(figsize=(10, 6))
sns.histplot(df['Sales'], bins=50, kde=True)
plt.title('Sales Distribution')
plt.xlabel('Sales')
plt.ylabel('Frequency')
plt.yscale('log')
plt.show()
```

### 6. Time Series Plot

```python
sales_year = df.groupby(df['Order Date'].dt.to_period('Y'))['Sales'].sum()

plt.figure(figsize=(12, 6))
plt.plot(sales_year.index.to_timestamp(), sales_year.values, marker='o', linestyle='-')
plt.title('Sales Past 4 Years')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.grid(True)
plt.tight_layout()
plt.show()
```

### 7. Seasonal Decomposition

```python
plt.figure(figsize=(12, 8))

plt.subplot(4, 1, 1)
plt.plot(df['Sales'], label='Original Time Series')
plt.legend()
```

### 8. Box Plots

```python
df.boxplot('Sales')
plt.title('Boxplot of Sales')
plt.ylabel('Sales')
plt.show()
```

###9. Sales Distribution by Category

```python
df.plot(x='Region', y='Sales', kind='scatter')
plt.title('Sales Vs Region')
plt.xlabel('Region')
plt.ylabel('Sales')
plt.show()
```

```python
df.plot(x='Sales', y='State', figsize=(12, 10), kind='scatter')
plt.title('Sales Vs State')
plt.xlabel('Sales')
plt.ylabel('State')
plt.show()
```

```python
df['Date'] = pd.to_datetime(df['Order Date'])
```

```python
# Extract year, month, and quarter from the ''Order Date' column
df['Year'] = df['Date'].dt.year
df['Month'] = df['Date'].dt.month
df['Quarter'] = df['Date'].dt.quarter

# Group by year, month, and quarter and sum the sales
sales_by_year = df.groupby('Year')['Sales'].sum()
sales_by_month = df.groupby(['Year', 'Month'])['Sales'].sum()
sales_by_quarter = df.groupby(['Year', 'Quarter'])['Sales'].sum()
```

```python
plt.figure(figsize=(12, 9))
plt.subplot(3, 1, 1)
sales_by_year.plot(kind='bar', color='skyblue')
plt.title('Total Sales by Year')
plt.xlabel('Year')
plt.ylabel('Sales')
```

```python
plt.figure(figsize=(12, 9))
plt.subplot(3, 1, 3)
sales_by_month.plot(kind='bar', color='red')
plt.title('Total Sales by Month')
plt.xlabel('Year-Month')
plt.ylabel('Sales')
```

```python
plt.figure(figsize=(12, 9))
plt.subplot(3, 1, 3)
sales_by_quarter.plot(kind='bar', color='green')
plt.title('Total Sales by Quarter')
plt.xlabel('Year-Quarter')
plt.ylabel('Sales')
```
