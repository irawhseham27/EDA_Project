# Final Findings:

1. **Make and Model Distribution:**
   - Maruti Suzuki is the most popular car make, followed by Hyundai, Kia, Tata, and Toyota.
   - The most popular car models are the Swift, Alto, Grand i10, Baleno, and Creta.

2. **Body Type and Fuel Type:**
   - Hatchbacks and SUVs are the most common body types.
   - Petrol is the most popular fuel type, followed by diesel and CNG.

3. **Safety Features:**
   - Child safety locks are present in almost all cars.
   - Most cars have rear parking sensors or cameras.
   - Navigation systems are becoming increasingly common.

4. **Pricing and Engine Specifications:**
   - The ex-showroom price of cars ranges from a few lakhs to several crores.
   - The displacement of engines ranges from 72 cc to 1950 cc.
   - The number of cylinders ranges from 1 to 8.

5. **Mileage and Warranty:**
   - The city mileage of cars ranges from 10 km/L to 30 km/L.
   - The basic warranty period for most cars is 3 years.

6. **Correlation Analysis:**
   - There is a strong positive correlation between ex-showroom price and displacement.
   - There is a moderate positive correlation between ex-showroom price and number of cylinders.
   - There is a weak positive correlation between city mileage and ex-showroom price.
